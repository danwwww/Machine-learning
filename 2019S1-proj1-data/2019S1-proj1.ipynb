```python
import csv
import math
```

```python
# This function should open a data file in csv, and transform it into a usable format 
def preprocess(name_of_csvFiles):
    
    #open the csv file
    with open(name_of_csvFiles,"r") as csvFiles:
        reader = csv.reader(csvFiles)
        
        #transorm the data into lists
        data = [row for row in reader]
                        
    csvFiles.close()
    
    print(data)
    return data

data = preprocess("anneal.csv")
data
```

```python
# This function should build a supervised NB model
def train(data):
    
    classes = {}
    cond_probs = []
    
    # find the count of each class
    for row in data:
        #put the data in a dictionary with class labels as keys and counts as values
        if row[-1] in classes:
            classes[row[-1]] += 1
        else:
            classes[row[-1]] = 1
    
    # initialize the list
    for row in data:
        if not cond_probs:
            #find the number of cloumns of the dataset and create a dictionary for each column
            for i in range(len(row)):
                cond_probs.append({})
                #find all types of class labels and create a dictionary for each label
                for label in classes.keys():
                    cond_probs[i][label] = {}
                    
        # count all the conditional probabilities
        for j in range(len(row)):
            class_label = row[-1]
            attributes = row[j]
            #count the number of each attributes
            for key in classes.keys():
                if attributes not in cond_probs[j][key]:
                    cond_probs[j][key][attributes] = 0
            if attributes in cond_probs[j][class_label]:
                cond_probs[j][class_label][attributes] += 1
            
            
    model = [list(classes.values()), cond_probs]
    return model

model = train(data)
model
```

```python
# This function should predict the class for an instance or a set of instances, based on a trained model 
def predict(model, data):
    classes = {}
    cond_probs = model[1]
    class_predict = []
    
    # find the count of each class
    for row in data:
        #put the data in a dictionary with class labels as keys and counts as values
        if row[-1] in classes:
            classes[row[-1]] += 1
        else:
            classes[row[-1]] = 1
    

    # count the probabilities for every class label
    for row in data:
        label_prob = {}
        
        for key in classes.keys():
            key_count = classes[key]
            prob_sum = 0
            
            for i in range(len(row)):
                if cond_probs[i][key][row[i]] != 0:
                    atbt_prob = cond_probs[i][key][row[i]]/key_count
                
                prob_sum += math.log(atbt_prob)
                
            key_prob = math.log(classes[key]) + prob_sum
            label_prob[key] = key_prob
            
        class_predict.append(max(label_prob, key = label_prob.get))
        
    return class_predict

prediction = predict(model, data)
prediction
```

```python
def evaluate(model, data):
    
    evaluation = dict([])
    classes = {}
    accuracy = 0
    total_presion = 0
    total_recall = 0
    
    
    # find the count of each class
    for row in data:
        #put the data in a dictionary with class labels as keys and counts as values
        if row[-1] in classes:
            classes[row[-1]] += 1
        else:
            classes[row[-1]] = 1
    
    for class_names in classes.keys():
        classes[class_names] = {}
        classes[class_names]["right"] = 0
        classes[class_names]["wrong"] = 0
  
    actual_class = list(row[-1] for row in data)
    predict_result = predict(model, data)
    
    # count right and wrong number of the preidict result  
    for i in range(len(actual_class)):
        if actual_class[i] == predict_result[i]:
            classes[actual_class[i]]["right"] += 1
        else:
            classes[predict_result[i]]["wrong"] += 1

    # evaluate the result of the prediction
    for result in classes.keys():
        
        TP = classes[result]["right"] 
        FP = classes[result]["wrong"]
        FN = 0
        
        for key in classes.keys():
            if key != result:
                FN += classes[key]["wrong"]
       
        #calculate the precision of each attribute
        accuracy += TP
        
        # calculate recall of each attribute
        if TP + FP != 0:
            presion = TP / (TP + FP)
        else:
            presion = 0
        total_presion += presion
        
        # calculate recall of each attribute
        if TP + FN != 0:
            recall = TP / (TP + FN)
        else:
            recall = 0
        total_recall += recall
        
    evaluation["Accuracy"] = accuracy/len(actual_class)
    evaluation["Presion"] = total_presion/len(model[0])
    evaluation["Recall"] = total_recall/len(model[0])
    
    return evaluation

evaluation = evaluate(model, data)
evaluation
```

```python
# This function should calculate the Information Gain of an attribute or a set of attribute, with respect to the class
def info_gain(data):
    
    cond_probs = model[1]    
    info_gain_list = []
    
    classes = {}
    # find the count of each class
    for row in data:
        #put the data in a dictionary with class labels as keys and counts as values
        if row[-1] in classes:
            classes[row[-1]] += 1
        else:
            classes[row[-1]] = 1
    
    total = 0
    root_entropy = 0
    
    #calculate the entropy of the root
    for class_label in classes.keys():
        total += classes[class_label]
        
    for class_label in classes.keys():
        prob =  classes[class_label] / total
        root_entropy += prob * math.log2(prob)
    root_entropy = -root_entropy
    print(root_entropy)
    
    
    for attributes_dict in cond_probs: 
        # put the count of instance value into a dictionary
        atbt_label_dict = {}
        for class_key in attributes_dict.keys():
            for atbt_key in attributes_dict[class_key].keys():
                if atbt_key not in atbt_label_dict.keys():
                    atbt_label_dict[atbt_key] = []
                atbt_label_dict[atbt_key].append(attributes_dict[class_key][atbt_key])
        
        # calculate mean information
        mean_info = 0           
        for keys in atbt_label_dict.keys():
            atbt_entropy = 0;
            for values in atbt_label_dict[keys]:
                prob_class = values/sum(atbt_label_dict[keys]);
                if (prob_class != 0):
                    atbt_entropy += prob_class * (math.log2(prob_class))
            mean_info +=  (-atbt_entropy) * (sum(atbt_label_dict[keys]) / sum(classes.values()))
    
        information_gain = root_entropy - mean_info
        info_gain_list.append(information_gain)  
    
    return info_gain_list

information_gain=info_gain(data)
information_gain
```
